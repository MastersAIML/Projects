```python
import numpy as np
import pandas as pd
import json
pd.options.display.float_format = '{:.2f}'.format

ratings = pd.read_csv("MyFoodRatings.csv")
ratings.head(30)
```

```python
data = ratings.drop(['Name'], axis=1)

U, s, VT = np.linalg.svd(data, full_matrices=True)
```
