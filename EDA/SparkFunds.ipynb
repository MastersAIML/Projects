```python
#Checkpoints-Part1

import numpy as np
import pandas as pd

# Load companies
companies = pd.read_csv("companies.txt", sep="\t",  encoding = "palmos")

# Load rounds2
rounds2 = pd.read_csv("rounds2.csv", encoding = "palmos")

rounds2.head()
```

```python
# unique companies in rounds2
# unique id is company_permalink
rounds2.company_permalink.str.lower().unique().size
```

```python
# unique companies in companies
# unique id is permalink
companies.permalink.str.lower().unique().size
```

```python
# column can be used as the unique key for each company
companies.info()
# permalink can be used
```

```python
# any companies in the rounds2 file which are not present in companies
intersection = np.intersect1d(rounds2.company_permalink.str.lower().unique(),companies.permalink.str.lower().unique()).size
rounds2unique = rounds2.company_permalink.str.lower().unique().size
print(intersection)
print(rounds2unique)
print(intersection == rounds2unique)
# No
```

```python
# Merge the two data frames so that all variables (columns) in the companies frame are added to the rounds2 data frame

# Make the unique key case insensitive for both the data frames
companies['permalink'] = companies['permalink'].str.lower()
rounds2['company_permalink'] = rounds2['company_permalink'].str.lower()

# merge the two frames
master_frame = pd.merge(rounds2, companies, left_on='company_permalink', right_on='permalink', how='left')

master_frame.head()

# How many observations are present in master_frame
master_frame.size
```

```python
#Checkpoints-Part2

#Cleaning of Data
master_frame.info()
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#Cleaning Date for raised_amount as its important to find the right investment type
master_frame = master_frame[~pd.isnull(master_frame['raised_amount_usd'])]
master_frame.info()
```

```python
#most representative value of the investment amount for each of the four funding types (venture, angel, seed, and private equity)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
by_funding_round_type = master_frame.groupby('funding_round_type')
by_funding_round_type = by_funding_round_type['raised_amount_usd'].mean().sort_values(ascending=False).reset_index()
print(by_funding_round_type)

#Convert to Million USD
by_funding_round_type.raised_amount_usd = by_funding_round_type.raised_amount_usd/1000000

# which investment type do you think is the most suitable for Spark Funds, wants to invest between 5 to 15 million USD per investment round
by_funding_round_type.loc[by_funding_round_type['funding_round_type'].isin(['venture', 'angel', 'seed', 'private_equity'])]
by_funding_round_type.loc[(by_funding_round_type['raised_amount_usd'] > 5) & (by_funding_round_type['raised_amount_usd'] < 15)]
# venture
```

```python
#Checkpoint 3: Country Analysis

#Spark Funds wants to see the top nine countries which have received the highest total funding (across ALL sectors for the chosen investment type)
#Chosen investment type is venture
venture_frame = master_frame.loc[(master_frame['funding_round_type'] == 'venture') & (master_frame['raised_amount_usd'] >= 5000000) & (master_frame['raised_amount_usd'] <= 15000000), :]
venture_frame.info()
```

```python
# Clean the country data. Founding_round_code is irrelevant here
round(100*(venture_frame.isnull().sum()/len(venture_frame.index)), 2)
venture_frame = venture_frame[~pd.isnull(venture_frame['country_code'])]
venture_frame.info()
```

```python
#Top 9 countries with highest funding in venture type
pd.set_option('display.float_format', lambda x: '%.2f' % x)
by_country_venture_df = venture_frame.groupby(['country_code']).raised_amount_usd.sum()
top_9_venture = by_country_venture_df.sort_values(ascending=False).head(9)
print(top_9_venture)
```

```python
# Checkpoint 4: Sector Analysis 1

# merged data frame with each primary sector mapped to its main sector (the primary sector should be present in a separate column)
sector_frame = pd.read_csv('mapping.csv')

#Extract the primary sector of each category list from the category_list column
venture_frame['primary_sector'] = venture_frame['category_list'].str.split('|',1).str[0]
cols = venture_frame.columns.tolist()
cols.insert(0, cols.pop(cols.index('primary_sector')))
venture_frame = venture_frame.reindex(columns= cols)
venture_frame.info()

#clean the data with no sectors
round(100*(venture_frame.isnull().sum()/len(venture_frame.index)), 2)
venture_frame = venture_frame[~pd.isnull(venture_frame['primary_sector'])]
venture_frame.info()
```

```python
#Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors (Note that ‘Others’ is also considered one of the main sectors)
sector_frame['primary_sector'] = sector_frame['category_list'].str.split('|',1).str[0]
sector_frame.info()
```

```python
# merge the two frames venture and mapping
venture_frame = pd.merge(venture_frame, sector_frame, on='primary_sector', how='left')
venture_frame.info()
```

```python
# Map the main_sector with primary_sector
def main_sector_fn(row):
    if row['Automotive & Sports'] == 1 :
        return 'Automotive & Sports'
    elif row['Blanks'] == 1 :
        return 'Blanks'
    elif row['Cleantech / Semiconductors'] == 1 :
        return 'Cleantech / Semiconductors'
    elif row['Entertainment'] == 1 :
        return 'Entertainment'
    elif row['Health'] == 1 :
        return 'Health'
    elif row['Manufacturing'] == 1 :
        return 'Manufacturing'
    elif row['News, Search and Messaging'] == 1 :
        return 'News, Search and Messaging'
    elif row['Others'] == 1 :
        return 'Others'
    elif row['Social, Finance, Analytics, Advertising'] == 1 :
        return 'Social, Finance, Analytics, Advertising'    
venture_frame['main_sector'] = venture_frame.apply(main_sector_fn, axis=1)
venture_frame.head()
```

```python
#Bringing the primary_sector and main_sector together
cols = venture_frame.columns.tolist()
cols.insert(1, cols.pop(cols.index('main_sector')))
venture_frame = venture_frame.reindex(columns = cols)
venture_frame.head()
```

```python
# Checkpoint 6: Plots

#A plot showing the fraction of total investments (globally) in venture, seed, and private equity, and the average amount of investment in each funding type

import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.float_format', lambda x: '%.2f' % x)
plt.figure(figsize=(10, 8))
by_funding_round_type = master_frame['raised_amount_usd'].groupby(master_frame['funding_round_type']).sum().reset_index()
by_funding_round_type['f_i'] = (by_funding_round_type['raised_amount_usd']/by_funding_round_type['raised_amount_usd'].sum())
by_funding_round_type_avg= master_frame['raised_amount_usd'].groupby(master_frame['funding_round_type']).mean().reset_index()
by_funding_round_type = pd.merge(by_funding_round_type,by_funding_round_type_avg, how='inner', left_on='funding_round_type', right_on='funding_round_type')
by_funding_round_type = by_funding_round_type.set_index('funding_round_type')
by_funding_round_type = by_funding_round_type.filter(items=['venture','seed','private_equity'], axis=0)
by_funding_round_type = by_funding_round_type.rename(index=str, columns={"raised_amount_usd_x":"total_investment", "f_i":"fraction_investment", "raised_amount_usd_y":"average_investment"})
by_funding_round_type['average_investment'] = by_funding_round_type['average_investment']/1000000
print(by_funding_round_type)
sns.set(style="white")
plt.subplot(1, 2, 1)
sns.barplot(x=by_funding_round_type.index, y=by_funding_round_type.fraction_investment, color="g", label='Fraction Investment').set(xlabel='Funding Round Type', ylabel='Fraction Investment')
plt.title('Fraction Investment by Funding Type')
plt.subplot(1, 2, 2)
sns.barplot(x=by_funding_round_type.index, y=by_funding_round_type.average_investment, color="b", label='Average Investment in Million USD').set(xlabel='Funding Round Type', ylabel='Average Investment')
plt.title('Average Investment in Million USD')
plt.show()
```

```python
# A plot showing the top 9 countries against the total amount of investments of funding type FT. 

plt.figure(figsize=(10, 8))
by_country = venture_frame['raised_amount_usd'].groupby(venture_frame['country_code']).sum().reset_index()
by_country['raised_amount_usd'] = by_country['raised_amount_usd']/1000000000
by_country = by_country.sort_values(by='raised_amount_usd', ascending=False).head(9)
barplot = sns.barplot(y='country_code', x='raised_amount_usd', data=by_country)
plt.title('Investment by Country')
barplot.set(xlabel='Amount Raised in Billion USD in Venture Investement',ylabel='Country Code')
print(by_country)
```

```python
# A plot showing the number of investments in the top 3 sectors of the top 3 countries on one chart (for the chosen investment type FT)

plt.figure(figsize=(10, 8))
by_country_sector = venture_frame['raised_amount_usd'].groupby([venture_frame.country_code,venture_frame.main_sector]).count().reset_index()
by_country_sector = by_country_sector.loc[(by_country_sector['country_code'] == 'USA') | (by_country_sector['country_code'] == 'GBR') | (by_country_sector['country_code'] == 'CAN'), :]
by_country_sector = by_country_sector.sort_values(by=['country_code','raised_amount_usd'], ascending=[1,0])
gbr = by_country_sector.loc[by_country_sector['country_code'] == 'GBR', :].head(3)
usa = by_country_sector.loc[by_country_sector['country_code'] == 'USA', :].head(3)
can = by_country_sector.loc[by_country_sector['country_code'] == 'CAN', :].head(3)
by_country_sector = pd.concat([usa, gbr, can])
by_country_sector = by_country_sector.rename(columns={"raised_amount_usd": "no_of_investments"})
barplot = sns.barplot(y='country_code', x='no_of_investments', hue="main_sector", data=by_country_sector)
barplot.set(xlabel='Number of investments in Venture Investement',ylabel='Country Code')
plt.title('Investment by top 3 sectors of top 3 Countries')
print(by_country_sector)
plt.show()
```
